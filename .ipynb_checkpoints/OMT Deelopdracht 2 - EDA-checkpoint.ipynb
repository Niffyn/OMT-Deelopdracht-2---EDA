##Benodigde data:

Om deze vraag te beantwoorden, combineren we twee datasets:

Dataset 1: Tweets over de Oekra√Øense oorlog, verzameld op 19 augustus 2022 tot heden. 

Dataset 2: Gegevens over Russische verliezen in militair materieel en personeel, verzameld van 25 januari 2022 tot heden.

## 2.1: Stappen

1- Sentimentanalyse: Voer een sentimentanalyse uit op de tweets om de emoties (positief, negatief, neutraal) met Natural Language Processing van elke tweet te coderen.

2- Activiteitsanalyse: Analyseer het aantal tweets per dag om pieken in activiteit te identificeren.

3- Correlatieanalyse: Onderzoek de correlatie tussen de dagelijkse Russische verliezen en zowel het sentiment als de activiteit op Twitter.

4-Botdetectie: Identificeer mogelijke botaccounts door te kijken naar kenmerken zoals de aanmaakdatum van het account en het aantal volgers. 

5- Botdetectie correlatatieanalyse: Bekijk of er een toename is in het aantal tweets van vermoedelijke botaccounts na significante Russische verliezen wat kan wijzen op geco√∂rdineerde desinformatiecampagnes.

## 2.2: Relevantie van de onderzoeksvraag

Dit onderzoek is maatschappelijk en wetenschappelijk relevant omdat het inzicht biedt in hoe militaire gebeurtenissen het sentiment van online communicatie kan be√Ønvloeden. Daarnaast kan het identificeren van mogelijke desinformatiecampagnes helpen bij het ontwikkelen van strategie√´n om de verspreiding van nepnieuws tegen te gaan, wat essentieel is voor het behoud van een ge√Ønformeerde samenleving. Wat dit onderzoek meerwaarde geeft is dat het de timing van desinformatiecampagnes kan identificeren en zo dus het plan van de Russische overheid kan aankaarten, wel moet er een innovatief/nieuw analysemodel gebruikt worden.

## 3.1: Nauwkeurigheid

De 'tweets' dataset is nauwkeurig, omdat het simpelweg een collectie is van tweets en informatie over de tweets. Voor 'pers_plus_equip' geldt dit ook, maar de data is verzameld door de Oekra√Øense overheid onder leiding van Petro Ivaniuk. De Oekra√Øense overheidsinstantie claimt 100% correct te zijn zonder schatting, maar dat betwijfel ik. Ze kunnen liegen, omdat ze een politiek doel hebben en de informatie van de Russische overheid hierover is niet makkelijk toegankelijk zonder spionage en hacking. Het moet wel een schatting zijn, maar hun level of confidence is hoog. Ze kunnen namelijk zelf de apparatuur en doden tellen. '
Uit de dataset blijkt dat bepaalde waarden, zoals aantallen vernietigde tanks en vliegtuigen, extreem hoog lijken met veel outliers en grote verandering in korte tijdsperiodes. Dit suggereert dat de Oekra√Øense overheid mogelijk een te optimistische schatting maakt. Ook ontbreekt een formele methode om de cijfers te verifi√´ren, waardoor de nauwkeurigheid slechts gedeeltelijk kan worden gegarandeerd. Ik beveel een externe, onafhankelijke validatie van de gegevens zou de nauwkeurigheid verbeteren.

## 3.2: Precisie

Voor 'tweets' is de data precies, omdat alle tekst van de tweet en de informatie over de tweets is beschreven. 'pers_plus_equip' is ook precies, want er wordt gewoon geteld hoeveel er van iets is. Wel wordt er NaN gebruikt en bij pers_plus_equip['submarines'] en pers_plus_equip['vehicle and fuel tanks'] soms floats gebruikt terwijl er geen halve tanks en onderzeeboten bestaan, maar er worden geen echte kommagetallen gebruikt. 1 is bijvoorbeeld geschreven als 1.0 (float). Ik zou het gebruik van gehele getallen in plaats van floats om de precisie van de dataset verbeteren aanbevelen.

## 3.3: Consistentie

Beide datasets zijn consistent. De tweets worden op dezelfde manier gerepresenteerd en het apparatuur/personeelsverlies wordt op dezelfde manier geteld. De categorie√´n van apparatuur zijn globaal genoeg voor duidelijke categorisering, ook wordt er geen uitzondering gemaakt van personeel. Om er 100% zeker van te zijn, moet er wel verder onderzoek gedaan worden of er per dag consistent gecategoriseerd wordt. Dit is de 'hidden consistency'. Aanbeveling: Verder onderzoek naar de dagelijkse categorisering kan helpen om consistentie op een dieper niveau te verifi√´ren van een onafhankelijke derde partij.

## 3.4: Compleetheid

Voor 'tweets' zijn alle kolommen relevant en hebben ze geen correlatie met elkaar, want het is tekstdata. 'pers_plus_equip' is ook compleet, want alle categorie√´n van personeel en apparatuur zijn distinct. Het tellen hiervan heeft niets met elkaar te maken. Bij 'pers_plus_equip' kan het ontbreken van detailinformatie, zoals de exacte bron van de gegevens of de manier waarop verliezen zijn geverifieerd, zorgen voor een incompleet beeld. Er zijn ook geen subcategorie√´n binnen bepaalde typen apparatuur, zoals onderscheid tussen verschillende modellen tanks of vliegtuigen. Bovendien ontbreken contextuele gegevens zoals gevechtslocaties of omstandigheden waarin de verliezen zijn opgetreden, wat de volledigheid van de dataset beperkt. Ik suggereer om extra van subcategorie√´n of meer gedetailleerde informatie toe te voegen om de volledigheid van de dataset verbeteren.

De 'tweets' data zijn observaties die niet bewerkt of geclassificeerd zijn. Dit kan dus niet foutief geclassificeerd of foutief gerepresenteerd zijn. Voor beide datasets is de data per tweet uniek, evenals de hoeveelheid personeel en apparatuur per dag. Dit geldt voor alle kolommen. Er hoeft geen controlevraag gesteld te worden, want dit zijn observaties. Er zijn geen duplicaties. Uniekheid kan op twee niveaus worden bekeken: horizontaal (duplicatie van kolommen) en verticaal (duplicatie van rijen). In beide datasets lijkt er geen sprake te zijn van horizontale redundantie, omdat elke kolom een specifieke en niet-overlappende informatiebron representeert. Ook is er geen sprake van verticale redundantie, aangezien observaties per tijdseenheid worden vastgelegd en niet dubbel voorkomen. Desondanks kan een controlemechanisme worden ge√Ømplementeerd om toekomstige duplicaties te voorkomen, vooral bij het samenvoegen van meerdere datasets. Aanbeveling: Het implementeren van geautomatiseerde checks voor duplicaties, vooral als er meerdere datasets gecombineerd moeten worden.

# Opdracht 4: Univariate Analyse

Voor de eerderbenoemde belangrijkste kolommen is er geen missende data. De gekozen vijf belangrijkste kolommen zijn:

tweets['retweetcount'] is vergeleken met andere kollomen minder gevarieerd door de grote hoeveelheden aan nullen. Dit kan zijn door bot accounts, of omdat er heel veel accounts zijn die niet zo veel bereik hebben en alleen tweeten voor een kleine doelgroep. De distributie is dus positief geskewed.

```python
# Gemiddelde, mediaan, modus, standaardafwijking en variantie voor tweets['retweetcount']
variabele = tweets['retweetcount']

gemiddelde = float(variabele.mean())
mediaan = variabele.median()
modus = variabele.mode().iloc[0]
standaardafwijking = float(variabele.std())
variantie = variabele.var()

print(f"Gemiddelde: {gemiddelde}")
print(f"Mediaan: {mediaan}")
print(f"Modus: {modus}")
print(f"Standaardafwijking: {standaardafwijking}")
print(f"Variantie: {variantie}")
```

```python
# Boxplot voor ARRIVAL_DELAY
sns.boxplot(data=tweets, x= 'retweetcount')
plt.title("Boxplot van tweets['retweetcount']")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(tweets['retweetcount'], kde=False, discrete=True, color="blue", edgecolor="black")
plt.xlim(0, int(gemiddelde + standaardafwijking) * 1.25)
plt.ylim(0, variabele.value_counts().max() * 1.1)  
plt.title("Verdeling van tweets['retweetcount']")
plt.show()
```

# Opdracht 5: Multivariate analyse

In een echt onderzoek zouden alle twitter datasets samengevoegd worden en het sentiment per tweet gemeten worden (bijvoorbeeld ). Er zou dan een gemiddeld sentiment nodig zijn per dag, samen met het gemiddeld bereik (hoeveel mensen deze berichten potentieel bereikt hebben doormiddel van een score aan de hand van followers, retweetcount etc.) en de meest voorkomende hashtag voor die dag. Dit moet gebeuren, omdat 'pers_plus_equip' per dag is beschreven in de dataset. In een echt onderzoek kan er dan correlatieonderzoek gedaan worden, maar omdat de tweets dataset te groot is ga ik er van uit dat elke rij in 'tweets' een gemiddelde tweet per dag is om toch correlatieonderzoek te doen. Dit heb ik samengevoegd in dataframe 'combined'.

```python
len(pers_plus_equip), len(tweets)
temp1 = pers_plus_equip.copy()
temp2 = tweets[:1102].copy()
index = [x for x in range(1102)]
temp1.index = index
temp2.index = index
temp1.index.name = 'Index'
temp2.index.name = 'Index'
combined = temp1.merge(temp2, how="inner", on="Index")
```

# Opdracht 6: Ethische analyse

In datasets zoals 'tweets' en 'pers_plus_equip' kunnen verschillende vormen van bias optreden die de betrouwbaarheid en objectiviteit van de gegevens be√Ønvloeden. Hieronder worden drie specifieke vormen van bias besproken, met nadruk op hoe deze zich kunnen manifesteren in de 'pers_plus_equip' dataset:

1- Propaganda (Confirmation Bias)

Overheden die betrokken zijn bij conflicten, zoals de Oekra√Øense overheid, kunnen geneigd zijn informatie te presenteren die hun eigen standpunten of politieke agenda ondersteunt. Dit kan leiden tot manipulatie van publieke opinie en kan ook invloed hebben op politieke beslissingen , waarbij bijvoorbeeld verliezen van de vijand worden overdreven en eigen verliezen worden geminimaliseerd om en externe steun te versterken. Een vergelijkbaar fenomeen is waargenomen in de media, waar selectieve berichtgeving kan leiden tot een vertekend beeld van een conflict. 

2- Selectiebias

Wanneer datasets voornamelijk zijn gebaseerd op bronnen van √©√©n partij in een conflict, zoals militaire rapporten van Oekra√Øne, bestaat het risico op selectiebias. Dit houdt in dat de verzamelde gegevens een onvolledig of eenzijdig beeld geven van de situatie, omdat ze mogelijk belangrijke informatie van andere partijen, zoals Russische bronnen of neutrale waarnemers, negeren. Dit is bij deze dataset het geval, omdat de Oekra√Øense overheid deze dataset bijhoudt.

3- Verificatiebias

In oorlogsomstandigheden is het vaak moeilijk om informatie onafhankelijk te verifi√´ren. Dit kan ertoe leiden dat onjuiste of overdreven cijfers over bijvoorbeeld verliezen worden verspreid zonder adequate bevestiging. Tegelijkertijd kunnen gegevens die de successen of sterktes van de tegenpartij (in dit geval Rusland) tonen, worden weggelaten of geminimaliseerd. Het rapporteren van misinformatie van verloren apparatuur en militair personneel werd al vaker gedaan in dit conflict (Radford et al., 2023). 


### De volgende type bias treden *niet* voor in 'pers_plus_equip':

1- Survivorship Bias
Survivorship bias treedt op wanneer alleen de gegevens van "overlevers" worden geanalyseerd, terwijl de rest buiten beschouwing wordt gelaten. Een klassiek voorbeeld is het analyseren van succesvolle bedrijven zonder te kijken naar mislukte ondernemingen. In de context van de 'pers_plus_equip' dataset is dit minder relevant, omdat de dataset juist gericht is op verliezen (zoals vernietigde apparatuur en doden). Hierdoor wordt niet alleen gekeken naar de "overlevers" maar expliciet naar de "verliezers" in het conflict.

2- Observer Bias
Observer bias ontstaat wanneer de verwachtingen van een waarnemer de waarnemingen be√Ønvloeden, zoals in psychologische experimenten of medische studies. In de 'pers_plus_equip' dataset is deze bias minder relevant als de gegevens worden verzameld via observaties met minimale menselijke interpertatie.
